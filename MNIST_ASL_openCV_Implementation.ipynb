# Creating the word dictionary with indices

```python
L=[]
for i in range(65,91):
  if i==90:
    continue
  else:
    c=chr(i)
    L.append(c) 
print(L)
```

```python
word_dict={}
for i in list(range(25)):
  word_dict[i]=L[i]
print(word_dict)
```

# Predicting with real-time video

```python
import cv2

video = cv2.VideoCapture(0) # accessing my webcam

while True:
    check,frame= video.read() #readig the images from the webcam
    frame=cv2.flip(frame,1) # flipping the image as it creates mirror images.
    
    frame=cv2.cvtColor(frame,cv2. COLOR_RGB2BGR) # converting the image into RGB from BGR as opencv reads the images in BGR format.
        
    frame_copy=frame.copy
    print(check)
    roi=frame[100:350,150:400] # Selecting a random small window out of the whole frame. The hand should be placed at this window.
    gray_frame= cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY) # converting the image into grayscale as the model accepts grayscaled images.

    
    cv2.imshow("gray_Frame", gray_frame) # showing the grayscaled image.
    
    
    gray_frame1 = cv2.resize(gray_frame, (28,28)) # resizing as the sahpe which is accepted by my model.
    
    
    blob_frame=cv2.dnn.blobFromImage(gray_frame1, scalefactor=1.0/255) # rescaling.
    #print(blob_frame.shape)
    
    gray_frame2 = np.reshape(blob_frame,(1,blob_frame.shape[2],blob_frame.shape[3],1)) # introducing the 4th dimension to make it compatible with keras.
    
    pred = model.predict(gray_frame2)# ending the processed image to the pretrained model.
    
    text= str(word_dict[np.argmax(pred)])+"         " + str(pred[0][np.argmax(pred)]) # creating the message to be shown on the screen.
    
    cv2.putText(frame,text,(170, 45), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2) # putting the text on the image.
    #print("******************************")
    #print(pred[0][np.argmax(pred)])
    cv2.imshow("Sign Detection", frame) # showing the image with the text.
    
    key = cv2.waitKey(1)

    if key ==ord("q"):
        break
    # loop will break when 'q' is pressed.

    
video.release()    # releasing the access to the webcam.
cv2.destroyAllWindows() # destroying the created windows.
```
